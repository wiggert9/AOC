```julia
function process_lines(lines)
    col_idranges = []
    changes=0
    for idrange in lines
        processed = false
        for (i,check_range) in enumerate(col_idranges)
            a,b = idrange[1], idrange[2]
            x,y = check_range[1], check_range[2]
            if x<=a && b<=y
                processed = true
                break
            elseif x<=a<=y && b>y
                col_idranges[i][2]=b
                processed = true
                changes+=1
                break
            elseif x>a && x<=b<=y
                col_idranges[i][1]=a
                processed = true
                changes+=1
                break
            elseif x>a && b>y
                col_idranges[i][1]=a
                col_idranges[i][2]=b
                processed = true
                changes+=1
                break
            end
        end
        if processed == false
            push!(col_idranges, idrange)
        end
    end
    return col_idranges,changes
end

        
```

```julia
changes=5
count = 0
while changes>0 && count<1000
    lines, changes = process_lines(lines)
    println(lines)
    println(changes)
    count+=1
end
```
