```julia
lines = readlines("inputb")
```

```julia
pointer = 50
counter=0
checker=50
for rot in lines
    dir = rot[1]
    dist = parse.(Int,rot[2:end])
    
    if dir == 'R'
        checker = mod((checker + dist),100)
        if pointer + dist >= 100
            while pointer + dist >=100
                counter +=1
                dist -= 100
            end
            pointer = pointer + dist
        else
            pointer = pointer + dist
        end

    elseif dir == 'L'
        checker = mod((checker - dist),100)
        if pointer - dist <0
            if pointer == 0
                counter-=1
            end
            while pointer - dist <0
                counter+=1
                dist -=100
            end
            pointer = pointer-dist
        else
            pointer = pointer-dist
        end  
        if pointer ==0
            counter+=1
        end
    end
    println(dir)
    println(parse.(Int,rot[2:end]))
    println(dist)
    println(pointer)
    println(checker)
    println(counter)
    println()    
end
print(counter)

```
