```julia
text = readlines("inputb")
line = split(text[1],"")
lineint = [parse(Int,num) for num in line]
```

```julia
numarr = []
for linecat in text
    line = split(linecat, "")
    lineint = [parse(Int,num) for num in line]
    push!(numarr,lineint)
end 
println(numarr)
println(numarr[1])
```

```julia
sum =0
for line in numarr
    maxnum=0
    maxind = 0
    N = length(line)
    for i in 1:(N-1)
        num = line[i]
        if num>maxnum 
            maxnum=num
            maxind = i
        end
    end

    secondmaxnum = 0
    for j in maxind+1:N
        num = line[j]
        if num>secondmaxnum
            secondmaxnum=num
        end
    end
    println(maxnum, secondmaxnum)
    sum+= 10*maxnum + secondmaxnum
end
println(sum)
```
