```julia
text = readlines("inputb")
lines = []
prodline=[]
println(text[end])
```

```julia
snps = sort(vcat(findall.(x -> x .== '*', text[end]), findall.(x -> x .== '+', text[end])))
println(snps)
columns = []
for j in 1:length(snps)
    column = []
    for (i,line) in enumerate(text[1:end])
        if j != length(snps)
            push!(column, line[snps[j]:snps[j+1]-2])
        else
            push!(column, line[snps[j]:end])
        end
    end
    push!(columns, column)
end
println(columns)
```

```julia
numars = []
for col in columns
    nums = []
    for (i,string) in enumerate(col[1:end-1])
        if i==1
            for char in string
                push!(nums, char)
            end
        else
            for (j, char) in enumerate(string)
                nums[j] *= char
            end
        end
    end
    push!(numars, nums)
end
println(numars)

for (i,nums) in enumerate(numars)
    for (j, num) in enumerate(nums)
        numars[i][j] = parse(Int,(replace(num, " " => "")))
    end
end
println(numars)
```

```julia
totalsum =0
for (i,nums) in enumerate(numars)
    if occursin("*", columns[i][end])
        num = prod(nums)
    elseif occursin("+", columns[i][end])
        num = sum(nums)
    end
    println(num)
    totalsum+=num
end
println(totalsum)
```
