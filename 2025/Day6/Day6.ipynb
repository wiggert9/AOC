```julia
text = readlines("inputb")
lines = []
prodline=[]
for (i,line) in enumerate(text[1:end])
    while occursin("  ", line)
        line = replace(line, "  " => " ")
    end
    line= rstrip(line, [' '])
    line= lstrip(line, [' '])
    if i == length(text)
        prodline = split(line," ")
    else
        numline = parse.(Int, split(line, " "))
        push!(lines, numline)
    end
end
print(prodline)
print(lines)
```

```julia
sum=0
for (i,operator) in enumerate(prodline)
    num = lines[1][i]
    for j in 2:length(lines)
        if operator == "+"
            num += lines[j][i]
        elseif operator == "*"
            num *= lines[j][i]
        end
    end
    println(num)
    sum+=num
end
println(sum)
```
