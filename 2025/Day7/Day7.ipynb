```julia
text = readlines("inputb")
beamset = Set(findfirst(==('S'), text[1]))
```

```julia
count=0
for line in text[2:end]
    newbeamset = Set()
    for beam in beamset
        if line[beam]== '^'
            push!(newbeamset, beam-1)
            push!(newbeamset, beam+1)
            count+=1
        else
            push!(newbeamset, beam)
        end
    end
    beamset = newbeamset
end
println(count)
```

```julia
text = readlines("inputa")
beamset = Dict(findfirst(==('S'), text[1]) => 1)
count=0
for line in text[2:end]
    newbeamset = Dict()
    for (key,value) in beamset
        if line[key]== '^'
            if key-1 in keys(newbeamset)
                newbeamset[key-1] +=value
            else
                newbeamset[key-1] = value
            end

            if key+1 in keys(newbeamset)
                newbeamset[key+1] += value
            else
                newbeamset[key+1] = value
            end
        else
            newbeamset[key] = value
        end
    end
    beamset = newbeamset
    x = zeros(length(line))
    y = collect(line)
    for (key,value) in beamset
        x[key] = char(value)
    end
    println(y)
    println(x)
end

count = sum(values(beamset))
println(count)
```
