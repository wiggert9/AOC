```julia
text = readlines("inputb")
grid = reshape([c == '@' ? 1 : 0 for row in text for c in row],
                 length(text), length(text[1]))'

```

```julia
count = 0
m,n=size(grid)
summ=0
for i in 1:m
    for j in 1:n
        if grid[i,j] == 1
            tpcount =0
            summ+=1
            for di in -1:1, dj in -1:1
                if di ==0 && dj == 0
                    continue
                end
                ii=i+di
                jj=j+dj
                if 1<=ii<=m && 1<=jj<=n
                    if grid[ii,jj]==1
                        tpcount+=1
                    end
                end
                if tpcount>=4
                    summ-=1
                    break
                end
            end
        end
    end
end
println(summ)
```

```julia
function process_grid(grid)
    count = 0
    m,n=size(grid)
    summ=0
    indices = []
    for i in 1:m
        for j in 1:n
            if grid[i,j] == 1
                tpcount =0
                summ+=1
                push!(indices, (i,j))
                for di in -1:1, dj in -1:1
                    if di ==0 && dj == 0
                        continue
                    end
                    ii=i+di
                    jj=j+dj
                    if 1<=ii<=m && 1<=jj<=n
                        if grid[ii,jj]==1
                            tpcount+=1
                        end
                    end
                    if tpcount>=4
                        summ-=1
                        pop!(indices)
                        break
                    end
                end
            end
        end
    end
    for tuple in indices
        i,j = tuple
        grid[i,j]=0
    end
    return grid,summ
end

```

```julia
function recursive(grid)
    totalsum=0
    summ=1
    count=0
    while summ>0 && count<100
        grid, summ = process_grid(grid)
        #println(grid)
        #println(summ)
        #println()
        totalsum+=summ
        count+=1
    end
    #println(totalsum)
    return totalsum
end

```
