```julia
text = readlines("inputb")
```

```julia
lines = split(text[1], ",")
```

```julia
count = 0
sum=0
for line in lines
    nums = parse.(Int,(split(line, "-")))
    for num in range(nums[1],nums[2])
        length = Int(floor(log10(num))) +1
        if (length % 2) == 0 
            id2 = Int(num % (10^(length/2)))
            id1 = Int((num - id2)/10^(length/2)) 
            #println("$num, $id1, $id2")
            if id1 == id2
                count+=1
                sum+=num
            end
        end
    end
end
println(count)
println(sum)
```

```julia
count = 0
sum=0
prevnum = -999
for line in lines
    nums = parse.(Int,(split(line, "-")))
    for num in range(nums[1],nums[2])
        length = Int(floor(log10(num))) +1 
        for i in 1:Int(floor(length/2))
            if num == prevnum
                continue
            end
            id = []
            num_copy = copy(num)
            for j in 0:Int(floor(length/i))-1
                index = 10^i
                new_id = Int(num_copy % (index))
                if log10(new_id)<i
                    new_id = repeat("0",(i-1))*string(new_id)
                end
                push!(id, new_id)
                num_copy = Int((num_copy-parse(Int,new_id))/(index))
            end
            if num_copy >0
                push!(id,num_copy)
            end
            if all(x -> x == id[1], id)
                count+=1
                sum+=num
                prevnum = num
                continue
            end
        end
    end
end
println(count)
println(sum)
```
