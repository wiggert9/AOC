```julia
text = readlines("inputb")
ind_lights = []
buttons = []
joltage = []
for (i,line) in enumerate(text)
    split_line = split(line, " ")
    new_buttons = []
    lightsize = length(split_line[1]) -2
    for (j, word) in enumerate(split_line)
        if j==1
            arr= zeros(Int,lightsize)
            for (k,char) in enumerate(word[2:end-1])
                if char == '#'
                    arr[k]=1
                end
            end
            push!(ind_lights, arr)

        elseif j == length(split_line)
            nums = parse.(Int, split(word[2:end-1], ","))
            push!(joltage, nums)
        else
            nums = parse.(Int, split(word[2:end-1], ","))
            new_nums = zeros(Int, lightsize)

            for num in nums
                new_nums[num+1] = 1
            end
            push!(new_buttons, new_nums)
        end
    end
    push!(buttons, new_buttons)
end
println(ind_lights)
println(buttons)
println(joltage)
```

```julia
using JuMP, HiGHS
function optimize_button_pressing(lights, buttons, joltage)
    d = length(lights)
    N = length(buttons)
    
    #println(lights)
    #println(buttons)
    model = Model(HiGHS.Optimizer)
    set_silent(model)
    @variable(model, k[1:N]>=0, Int)
    @objective(model, Min, sum(k))
    @variable(model, m[1:d], Int)
    for j in 1:d
        @constraint(model, sum(k[i]*buttons[i][j] for i in 1:N) == joltage[j])
    end

    optimize!(model)
    return sum(value.(k))
end

```

```julia
summ = 0
for i in 1:length(ind_lights)
    lights = ind_lights[i]
    btns = buttons[i]
    jltge = joltage[i]
    result = optimize_button_pressing(lights, btns, jltge)
    #println("Minimum button presses for light set $i: $result")
    summ+=result
end
println(summ)
```
