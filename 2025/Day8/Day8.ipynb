```julia
text = readlines("inputb")
points =[]
for line in text
    numline = parse.(Int, split(line, ","))
    push!(points, numline)
end
println(points)
println(length(points))
```

```julia
dists = zeros(length(points), length(points))
for (i,point1) in enumerate(points)
    for (j,point2) in enumerate(points)
        if i==j
            dist =9999999999999999
        else
            dist = sum((point1.-point2).^2)
        end
        dists[i,j] = dist
    end
end
```

```julia
circuits = Dict()
N=1000
connection_count = 0
circuitnum=1
while connection_count <N
    nextcon = minimum(dists)
    p = findfirst(x -> x == nextcon, dists)
    dists[p] = 1e16
    dists[p[2],p[1]] = 1e16
    if !(p[1] in keys(circuits)) && !(p[2] in keys(circuits))
        circuits[p[1]] = circuitnum
        circuits[p[2]] = circuitnum
        circuitnum+=1
        connection_count+=1
    elseif (p[1] in keys(circuits)) && !(p[2] in keys(circuits))
        circuits[p[2]] = circuits[p[1]]
        connection_count+=1
    elseif !(p[1] in keys(circuits)) && (p[2] in keys(circuits))
        circuits[p[1]] = circuits[p[2]]
        connection_count+=1
    elseif (p[1] in keys(circuits)) && (p[2] in keys(circuits))
        if circuits[p[1]] != circuits[p[2]]
            p2circ = circuits[p[2]]
            for (key, value) in circuits
                if value == p2circ
                    circuits[key] = circuits[p[1]]
                end
            end
            connection_count+=1
        else
            connection_count+=1
            #println("Skipped one!")
        end
    end
end
println(circuits)
```

```julia
arr = zeros(Int,maximum(values(circuits)))
for (key,value) in circuits
    arr[value] += 1
end
println(arr)
top3 = partialsort(arr, 1:3, rev=true)
total = prod(top3)
```
